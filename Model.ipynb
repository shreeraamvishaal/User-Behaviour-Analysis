## Creating the Dependent and Independent variables

```python
X = dataset.drop(columns="enrolled")
y = dataset["enrolled"]
```

## Creating the training and testing datasets

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size = 0.2,
                                                    random_state = 0)
```

## Saving and then deleting user ids

```python
train_identity = X_train['user']
X_train = X_train.drop(columns = ['user'])
test_identity = X_test['user']
X_test = X_test.drop(columns = ['user'])
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train2 = pd.DataFrame(sc_X.fit_transform(X_train))
X_test2 = pd.DataFrame(sc_X.transform(X_test))
X_train2.columns = X_train.columns.values
X_test2.columns = X_test.columns.values
X_train2.index = X_train.index.values
X_test2.index = X_test.index.values
X_train = X_train2
X_test = X_test2
```

## Model: L1 Regularization Model

```python
from sklearn.linear_model import LogisticRegression

# Create the Logistic Regression model with L1 regularization
classifier = LogisticRegression(random_state=0, penalty='l1', solver='liblinear')

# Fit the model
classifier.fit(X_train, y_train)
```

## Making predictions

```python
y_pred = classifier.predict(X_test)
```

## Checking the results

```python
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, precision_score, recall_score
cm = confusion_matrix(y_test, y_pred)
acc = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred) # tp / (tp + fp)
recall = recall_score(y_test, y_pred) # tp / (tp + fn)
f1 = f1_score(y_test, y_pred)
```

```python
print('Accuracy:',acc)
print('Precision: ',round(precision,4))
print('Recall_score:',round(recall,4))
print('f1_score:',round(f1,4))
```

```python
#Checking the confusion matrix
cm
```

## Validating the accuracy using kfold

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = classifier, X = X_train, y = y_train, cv = 10)
print("SVM Accuracy: %0.3f (+/- %0.3f)" % (accuracies.mean(), accuracies.std() * 2))
```

## Analysing the coefficients

```python
pd.concat([pd.DataFrame(dataset.drop(columns = 'user').columns, columns = ["features"]),
           pd.DataFrame(np.transpose(classifier.coef_), columns = ["coef"])
           ],axis = 1)
```

## Improving the model by using Parameter Tuning(Grid Search)

```python
from sklearn.model_selection import GridSearchCV
C = [0.001,0.01,0.1,0.5, 0.9, 1, 2, 5, 10, 100, 1000]
penalty = ['l1','l2']
parameters = dict(C = C,penalty = penalty)
grid_search = GridSearchCV(estimator = classifier,
                           param_grid = parameters,
                           scoring = 'accuracy',
                           cv = 10,
                           n_jobs = -1)
grid_search = grid_search.fit(X_train, y_train)
best_accuracy = grid_search.best_score_
best_parameters = grid_search.best_params_
print("Best Accuracy: {:.2f} %".format(best_accuracy*100))
print("Best Parameters:", best_parameters)
```

## Improving model by using tuned parameters

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0, penalty = 'l2', C=0.01)
classifier.fit(X_train, y_train)
```

## Making predictions

```python
y_pred = classifier.predict(X_test)
```

## Checking results

```python
cm = confusion_matrix(y_test, y_pred)
acc = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred) # tp / (tp + fp)
recall = recall_score(y_test, y_pred) # tp / (tp + fn)
f1 = f1_score(y_test, y_pred)
```

```python
print('Accuracy:',acc)
print('Precision: ',round(precision,4))
print('Recall_score:',round(recall,4))
print('f1_score:',round(f1,4))
cm
```

## Validating the accuracy using kfold

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = classifier, X = X_train, y = y_train, cv = 10)
print("SVM Accuracy: %0.3f (+/- %0.3f)" % (accuracies.mean(), accuracies.std() * 2))
```
